```python
def show_grayscale(img):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 3))
    ax1.imshow(img)
    ax2.imshow(np.round(np.mean(img, axis=-1)), cmap=plt.cm.gray)
    
    plt.show()
```
